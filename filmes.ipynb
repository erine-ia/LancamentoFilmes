```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
filmes = pd.read_json('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt')
filmes.head()
```

```python
#renomeando variáveis
filmes = filmes.rename(columns ={'Film':'filme','Year':'ano','Genre':'genero','Worldwide_Gross_M':'bilheterias U$', 'Budget_M':'Orçamento U$' })
filmes
```

```python
#tamanho do dataframe (linhas, colunas)
filmes.shape
```

```python
#tipos de dados
filmes.dtypes
```

```python
## Apenas valores dos dados da coluna específica 'Genre', sendo valores únicos
filmes['genero'].unique()
```

```python
#agrupamento com contagem de números de filmes em cada ano
filmes.groupby('ano')['filme'].nunique()
```

```python
#agrupamento total de bilheteria por gênero
filmes.groupby('genero')['bilheterias U$'].sum()
```

```python
#agrupamento total de bilheteria por gênero
filmes.groupby('genero')['Orçamento U$'].sum()
```

```python
#menor orçamento e bolheterias em milhões
men_orc = filmes['Orçamento U$'].min()
men_bil = filmes['bilheterias U$'].min()
print('menor orçamento: ', men_orc,'milhões\nmenor bilheteria: ', men_bil,'milhões')
```

```python
#maior orçamento e bileterias em milhões
mai_orc = filmes['Orçamento U$'].max()
mai_bil = filmes['bilheterias U$'].max()

print('maior orçamento: ', mai_orc,'milhões\nmaior bilheteria: ', mai_bil,'milhões')
```

```python
#visão geral
sns.pairplot(filmes)
```

```python
#Qual o filme com maior orçamento e com menor orçamento?


plt.figure(figsize = (10, 10))
plt.title('Orçamento de 30 filmes lançados entre 2007 e 2011')
plt.xlabel('Orçamento')
sns.set(style = 'whitegrid')
sns.barplot(y = 'filme',
            x = 'Orçamento U$',
            palette = 'inferno',
            data = filmes, 
           
           )
plt.xlabel('U$ Milhões')
plt.ylabel('Filmes')
plt.show()
```

- O filme com maior orçamento foi **Piratas do Caribe: No fim do Mundo** com 300 milhões de dólares, em 2007, ano em que os filmes de ação tiveram a maior média de orçamento. 
- O filme com menor orçamento foi **A Saga Crepúsculo: Lua Nova** com 50 milhões de dólares, em 2009. É o único filme do gênero entre os 30 analisados no ano de 2009. Entre os anos de 2007 a 2011, dos 30 filmes analisados, somente dois são do gênero Drama. As duas produções do gênero pertencem à Saga Crepúsculo, e possuem os menores orçamentos.

```python
#Qual o filme com maior bilheteria e com menor bilheteria?



plt.figure(figsize = (10, 10))
plt.title('Bilheteria de 30 filmes lançados entre 2007 e 2011')

sns.set(style = 'whitegrid')
sns.barplot(y = 'filme',
            x = 'bilheterias U$',
            
            data = filmes, 
           
           )
plt.xlabel('U$ Milhões')
plt.ylabel('Filmes')
plt.show()
```

- O filme de maior bilheteria entre 2007 e 2011, foi Avatar com 2.84 Bilhões de dólares. Número responsável por elevar a média de bilheterias para o gênero de ação no ano de 2009, único ano em que o gênero teve a maior média de bilheteria em relação a todos os outros. 

- O laçamento de menor bilheteria foi de Ratatouille em 2007, mesmo ano em que foi lançado 'Shrek Terceiro'. A animação obteve 621 milhões. Nesse ano, o gênero de Animação teve media de bilheterias abaixo de outros gêneros analisados. 


```python

# Em que anos foram feitos menos filmes de ação?
# E em que ano foram feitos menos filmes de animação?

plt.figure(figsize = (15, 6))
plt.title('Quantidade de filmes por gênero entre 2007 e 2011')
plt.xlabel('Orçamento')
sns.set(style = 'darkgrid')
sns.countplot(x = 'ano',
              hue= 'genero',
              data = filmes, 
              palette = 'gist_rainbow_r',
              edgecolor= '0',
              
              
           
           )
plt.xlabel('Anos')
plt.ylabel(' Quantidade de Filmes')

plt.show()
```

- Os anos de 2008 e 2010 foram os anos com menos lançamentos de filmes de Ação. 
- Os anos de 2008 e 2011 foram os anos com menos lançamentos de filmes de Animação



```python

# Em que ano os filmes de ação tiveram maior média de orçamento? 
# E em que ano os filmes de animação tiveram a maior média de orçamento?


plt.figure(figsize = (15, 6))
plt.title('Média de orçamento por gênero em cada ano')
plt.xlabel('Orçamento')
sns.set(style = 'darkgrid')
sns.barplot(x = 'ano',
            y = 'Orçamento U$',
            hue= 'genero',
            data = filmes, 
            palette = 'gist_rainbow_r',
            edgecolor= '0',
            ci = None
              
              
           
           )
plt.xlabel('Anos')
plt.ylabel('Média de Orçamento U$')

plt.show()


```

- Os filmes de ação tiveram a maior média de orçamento em 2007. 
- 2008 foi o ano em que a média de orçamento dos filmes de animação foi maior.
- Nos anos de 2008 e 2010 a media de orçamento dos filmes de animação superou a média dos demais gêneros.




```python
plt.figure(figsize = (15, 6))
plt.title('Média de Bilheteria por gênero em cada ano')

sns.set(style = 'darkgrid')
sns.barplot(x = 'ano',
            y = 'bilheterias U$',
            hue= 'genero',
            data = filmes, 
            palette = 'afmhot',
            edgecolor= '0',
            ci = None
              
              
           
           )
plt.xlabel('Anos')
plt.ylabel('Media de Bilheterias')

plt.show()
```
