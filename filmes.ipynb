```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
filmes = pd.read_json('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt')
filmes.head()
```

```python
#renomeando vari√°veis
filmes = filmes.rename(columns ={'Film':'filme','Year':'ano','Genre':'genero','Worldwide_Gross_M':'bilheterias U$', 'Budget_M':'Or√ßamento U$' })
filmes
```

```python
#Qual o filme com maior or√ßamento e com menor or√ßamento?


plt.figure(figsize = (10, 10))
plt.title('Or√ßamento de 30 filmes lan√ßados entre 2007 e 2011')
plt.xlabel('Or√ßamento')
sns.set(style = 'whitegrid')
sns.barplot(y = 'filme',
            x = 'Or√ßamento U$',
            palette = 'inferno',
            data = filmes, 
           
           )
plt.xlabel('U$ Milh√µes')
plt.ylabel('Filmes')
plt.show()
```

- O filme com maior or√ßamento foi **Piratas do Caribe: No fim do Mundo** com 300 milh√µes de d√≥lares, em 2007, ano em que os filmes de a√ß√£o tiveram a maior m√©dia de or√ßamento. 
- O filme com menor or√ßamento foi **A Saga Crep√∫sculo: Lua Nova** com 50 milh√µes de d√≥lares, em 2009. √â o √∫nico filme do g√™nero entre os 30 analisados no ano de 2009. Entre os anos de 2007 a 2011, dos 30 filmes analisados, somente dois s√£o do g√™nero Drama. As duas produ√ß√µes do g√™nero pertencem √† Saga Crep√∫sculo, e possuem os menores or√ßamentos.

```python
#Qual o filme com maior bilheteria e com menor bilheteria?



plt.figure(figsize = (10, 10))
plt.title('Bilheteria de 30 filmes lan√ßados entre 2007 e 2011')

sns.set(style = 'whitegrid')
sns.barplot(y = 'filme',
            x = 'bilheterias U$',
            
            data = filmes, 
           
           )
plt.xlabel('U$ Milh√µes')
plt.ylabel('Filmes')
plt.show()
```

- O filme de maior bilheteria entre 2007 e 2011, foi Avatar com 2.84 Bilh√µes de d√≥lares. N√∫mero respons√°vel por elevar a m√©dia de bilheterias para o g√™nero de a√ß√£o no ano de 2009, √∫nico ano em que o g√™nero teve a maior m√©dia de bilheteria em rela√ß√£o a todos os outros. 

- O la√ßamento de menor bilheteria foi de Ratatouille em 2007, mesmo ano em que foi lan√ßado 'Shrek Terceiro'. A anima√ß√£o obteve 621 milh√µes. Nesse ano, o g√™nero de Anima√ß√£o teve media de bilheterias abaixo de outros g√™neros analisados. 


```python

# Em que anos foram feitos menos filmes de a√ß√£o?
# E em que ano foram feitos menos filmes de anima√ß√£o?

plt.figure(figsize = (15, 6))
plt.title('Quantidade de filmes por g√™nero entre 2007 e 2011')
plt.xlabel('Or√ßamento')
sns.set(style = 'darkgrid')
sns.countplot(x = 'ano',
              hue= 'genero',
              data = filmes, 
              palette = 'gist_rainbow_r',
              edgecolor= '0',
              
              
           
           )
plt.xlabel('Anos')
plt.ylabel(' Quantidade de Filmes')

plt.show()
```

- Os anos de 2008 e 2010 foram os anos com menos lan√ßamentos de filmes de A√ß√£o. 
- Os anos de 2008 e 2011 foram os anos com menos lan√ßamentos de filmes de Anima√ß√£o



```python

# Em que ano os filmes de a√ß√£o tiveram maior m√©dia de or√ßamento? 
# E em que ano os filmes de anima√ß√£o tiveram a maior m√©dia de or√ßamento?


plt.figure(figsize = (15, 6))
plt.title('M√©dia de or√ßamento por g√™nero em cada ano')
plt.xlabel('Or√ßamento')
sns.set(style = 'darkgrid')
sns.barplot(x = 'ano',
            y = 'Or√ßamento U$',
            hue= 'genero',
            data = filmes, 
            palette = 'gist_rainbow_r',
            edgecolor= '0',
            ci = None
              
              
           
           )
plt.xlabel('Anos')
plt.ylabel('M√©dia de Or√ßamento U$')

plt.show()


```

- Os filmes de a√ß√£o tiveram a maior m√©dia de or√ßamento em 2007. 
- 2008 foi o ano em que a m√©dia de or√ßamento dos filmes de anima√ß√£o foi maior.
- Nos anos de 2008 e 2010 a media de or√ßamento dos filmes de anima√ß√£o superou a m√©dia dos demais g√™neros.
